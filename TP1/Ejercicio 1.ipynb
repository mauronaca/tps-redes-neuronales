```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.pyplot import imshow
from os import listdir
from PIL import Image
```

```python
images_path = './Imagenes'
imagenes = {}

for imagen in listdir(images_path):
    imagenes[imagen.split('.')[0]] = Image.open(images_path + '/' + imagen)

Np = len(imagenes)
    
plt.figure()
f, axarr = plt.subplots(1 , Np, figsize = (100,100))
i = 0
for imagen in imagenes: 
    axarr[i].imshow(imagenes[imagen])
    i += 1
```

```python
paloma = imagenes['paloma']
quijote = imagenes['quijote'] 
torero = imagenes['torero']

Np1 = 3

(ancho, largo) = paloma.size
N = ancho * largo # Cantidad de neuronas
```

```python
# Red para trs imagenes de mismo tamaño paloma, quijote y torero
px_paloma = (list(paloma.getdata())) 
px_torero = (list(torero.getdata()))
px_quijote = (list(quijote.getdata()))

# Matriz con los patrones, que serian 3. Tiene 2700 pixeles/filas (2700,3)
P = np.array([px_paloma, px_quijote, px_torero]).reshape(2700, 3)

# Transformo los pixeles a +1 blanco o -1 negro
# 255 -> blanco -> +1 , 0 -> negro -> -1 
P[np.where(P == 255)] = 1
P[np.where(P == 0)] = -1

# Matriz de sinapsis 
W = np.dot(P, P.T) - 3 * np.identity(2700)
print(W)

# Ejecutar la red
h = np.sign(np.dot(W, P[:,0]))

#with np.printoptions(threshold=np.inf):
#    print(salida-P[:,0])
#    print(salida-P[:,1])
#    print(salida-P[:,2])
```
